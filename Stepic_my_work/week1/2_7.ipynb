```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import cross_val_score
%matplotlib inline
from sklearn.model_selection import train_test_split
from graphviz  import Source
from IPython.display import SVG, display, HTML
style = "<style>svg{width: 50% !important; height: 50% !important;} </style>" 
```

```python
df = pd.read_csv('C:\\Users\\Gazauto\\Downloads\\train_data_tree.csv')
df.head()
```

```python
df.shape()
```

```python
X = df.drop(['num'], axis=1)
```

```python
y = df['num']
```

```python
X_train, y_train = X, y
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
parameters = {'criterion': ['entropy'], 'max_depth': range(1, 100)}
```

```python
grid_search_cv_clf = GridSearchCV(clf, parameters, cv=5)
```

```python
grid_search_cv_clf.fit(X_train, y_train)
```

```python
grid_search_cv_clf.best_params_
```

```python
best_clf = grid_search_cv_clf.best_estimator_
```

```python
# best_clf.score(X_test, y_test)
```

```python
tree.plot_tree(best_clf, filled=True)
```

```python
plt.figure(figsize=(40, 20),dpi=80)
p = tree.plot_tree(best_clf, fontsize=18,filled=True,feature_names=list(X))
```
